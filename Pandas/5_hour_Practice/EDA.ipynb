```python
# Quasi-identifiers
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=0.5)
sel.fit(food3)
```

```python
len(food3.columns[sel.get_support()])
```

```python
quasai_con = [col for col in food3.columns if col not in food3.columns[sel.get_support()]]
```

```python
len(quasai_con)
```

```python
foo = food3.drop(quasai_con,axis=1)
```
